```python
import pandas as pd
import plotly.express as px
from ipywidgets import interact, widgets

# Load the data
data_path = os.path.join(os.getcwd(), 'mock_methods_data.csv')
data = pd.read_csv(data_path)

# Convert loss_function from string to list (if saved as a string in CSV)
data['loss_function'] = data['loss_function'].apply(eval)

# Function to plot loss
def plot_loss(selected_meth1, selected_meth2, selected_meth3, selected_meth4):
    # Filter the row that matches the selected combination
    filtered_row = data[
        (data['meth1'] == selected_meth1) &
        (data['meth2'] == selected_meth2) &
        (data['meth3'] == selected_meth3) &
        (data['meth4'] == selected_meth4)
    ]
    
    if filtered_row.empty:
        print("No matching data for the selected combination of methods.")
        return
    
    # Extract the loss function
    loss_function = filtered_row['loss_function'].iloc[0]
    
    # Create a DataFrame for plotting
    plot_data = pd.DataFrame({'Epoch': range(1, len(loss_function) + 1), 'Loss': loss_function})
    
    # Create the plot
    fig = px.line(
        plot_data,
        x='Epoch',
        y='Loss',
        title="Training Loss Over Epochs",
        labels={'Epoch': 'Epoch', 'Loss': 'Loss'},
    )
    fig.show()

# Create interactive widgets
meth1_selector = widgets.Checkbox(value=False, description='meth1')
meth2_selector = widgets.Checkbox(value=False, description='meth2')
meth3_selector = widgets.Checkbox(value=False, description='meth3')
meth4_selector = widgets.Checkbox(value=False, description='meth4')

# Interactive function
@interact(
    meth1=meth1_selector,
    meth2=meth2_selector,
    meth3=meth3_selector,
    meth4=meth4_selector
)
def update_plot(meth1, meth2, meth3, meth4):
    plot_loss(meth1, meth2, meth3, meth4)
```

```python
df = pd.read_csv('info111.csv')
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
def transform_smotenc(value):
    # Check if the value is a string and represents a valid number
    if isinstance(value, str):
        try:
            # Try converting to float
            return float(value)
        except ValueError:
            # If conversion fails (e.g., it's 'False' or any invalid string), return the original value
            return value
    return value  # If it's already a number or False, return it as is

# Apply the transformation to the SMOTENC column
df['SMOTENC'] = df['SMOTENC'].apply(transform_smotenc)
df['Dropout'] = pd.to_numeric(df['Dropout'], errors='coerce')
```

```python
import pandas as pd
import plotly.graph_objects as go
from ipywidgets import interact, widgets
import base64
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

# Generate base64 placeholder image data
def create_base64_placeholder():
    from PIL import Image
    from io import BytesIO
    img = Image.new('RGB', (100, 100), color=(255, 0, 0))  # Create a red square
    buffer = BytesIO()
    img.save(buffer, format="PNG")
    return base64.b64encode(buffer.getvalue()).decode()

# Replace mock data with valid base64 strings
# data = pd.DataFrame({
#     'isol_forest': [True, False, True, False],
#     'iqr': [True, True, False, False],
#     'box_cox': [True, False, True, False],
#     'robust_scaler': [True, False, True, False],
#     'normalization': ['False', 'normalization', 'standardization', 'False'],
#     'SMOTENC': ['False', 0.4, 0.6, 0.8],
#     'one_hot': ['label_enc', 'one_hot', 'label_enc', 'one_hot'],
#     'Dropout': [0.00, 0.15, 0.25, 0.4],
#     'BatchNorm': [True, False, True, False],
#     'history': [
#         {'val_accuracy': [0.7, 0.75, 0.8], 'val_auc': [0.8, 0.85, 0.88], 'val_loss': [0.6, 0.5, 0.4],
#          'accuracy': [0.65, 0.7, 0.78], 'auc': [0.78, 0.82, 0.87], 'loss': [0.65, 0.55, 0.45]},
#         {'val_accuracy': [0.65, 0.7, 0.72], 'val_auc': [0.75, 0.78, 0.8], 'val_loss': [0.7, 0.6, 0.5],
#          'accuracy': [0.6, 0.65, 0.7], 'auc': [0.7, 0.75, 0.78], 'loss': [0.75, 0.65, 0.6]},
#         {'val_accuracy': [0.8, 0.82, 0.85], 'val_auc': [0.88, 0.9, 0.92], 'val_loss': [0.5, 0.45, 0.4],
#          'accuracy': [0.75, 0.8, 0.85], 'auc': [0.85, 0.9, 0.92], 'loss': [0.5, 0.45, 0.4]},
#         {'val_accuracy': [0.7, 0.72, 0.74], 'val_auc': [0.8, 0.82, 0.84], 'val_loss': [0.65, 0.6, 0.55],
#          'accuracy': [0.68, 0.7, 0.72], 'auc': [0.78, 0.8, 0.82], 'loss': [0.7, 0.65, 0.6]}
#     ],
#     't-SNE': [create_base64_placeholder(), create_base64_placeholder(),
#               create_base64_placeholder(), create_base64_placeholder()],
#     'confusion_matrix': [create_base64_placeholder(), create_base64_placeholder(),
#                          create_base64_placeholder(), create_base64_placeholder()]
# })

data = df


# Decode base64 string to display as an image
def decode_base64_image(encoded_str):
    decoded = base64.b64decode(encoded_str)
    image = Image.open(BytesIO(decoded))
    return image

# Function to plot history metrics
def plot_history(history):
    # Extract the validation metrics from history
    val_loss = history['val_loss']
    train_loss = history['loss']
    
    val_accuracy = history['val_accuracy']
    train_accuracy = history['accuracy']
    
    val_auc = history['val_auc']
    train_auc = history['auc']

    # Plot learning curves
    plt.figure(figsize=(12, 6))

    # Plot Validation Loss
    plt.subplot(1, 3, 1)
    plt.plot(val_loss, label='Validation Loss', color='red')
    plt.plot(train_loss, label='Train Loss', color='black')
    plt.title('Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot Validation Accuracy
    plt.subplot(1, 3, 2)
    plt.plot(val_accuracy, label='Validation Accuracy', color='blue')
    plt.plot(train_accuracy, label='Train Accuracy', color='black')
    plt.title('Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot Validation AUC
    plt.subplot(1, 3, 3)
    plt.plot(val_auc, label='Validation AUC', color='green')
    plt.plot(train_auc, label='Train AUC', color='black')
    plt.title('Validation AUC')
    plt.xlabel('Epochs')
    plt.ylabel('AUC')
    plt.legend()

    #Show all plots
    plt.tight_layout()
    plt.show()

# Function to plot t-SNE and confusion matrix
def plot_images(tsne_encoded, cm_encoded):
    tsne_image = decode_base64_image(tsne_encoded)
    cm_image = decode_base64_image(cm_encoded)
    
    # Display images side-by-side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(tsne_image)
    axes[0].set_title("t-SNE")
    axes[0].axis("off")
    
    axes[1].imshow(cm_image)
    axes[1].set_title("Confusion Matrix")
    axes[1].axis("off")
    
    plt.tight_layout()
    plt.show()

# Interactive plotting function
def interactive_plot(isol_forest, iqr, box_cox, robust_scaler, normalization, SMOTENC, one_hot, Dropout, BatchNorm):
    # Filter data for the selected options
    filtered_row = data[
        (data['isol_forest'] == isol_forest) &
        (data['iqr'] == iqr) &
        (data['box_cox'] == box_cox) &
        (data['robust_scaler'] == robust_scaler) &
        (data['normalization'] == normalization) &
        (data['SMOTENC'] == SMOTENC) &
        (data['one_hot'] == one_hot) &
        (data['Dropout'] == Dropout) &
        (data['BatchNorm'] == BatchNorm)
    ]
    
    if filtered_row.empty:
        print("No matching data for the selected combination.")
        return
    
    # Extract history and plot metrics
    history = filtered_row['history'].iloc[0]
    plot_history(history)
    
    # Extract and plot images
    tsne_encoded = filtered_row['t-SNE'].iloc[0]
    cm_encoded = filtered_row['confusion_matrix'].iloc[0]
    plot_images(tsne_encoded, cm_encoded)

# Create interactive widgets for each method
isol_forest_widget = widgets.Checkbox(value=False, description='isol_forest')
iqr_widget = widgets.Checkbox(value=False, description='iqr')
box_cox_widget = widgets.Checkbox(value=False, description='box_cox')
robust_scaler_widget = widgets.Checkbox(value=False, description='robust_scaler')
normalization_widget = widgets.Dropdown(options=['False', 'normalization', 'standardization'], value='False', description='normalization')
SMOTENC_widget = widgets.Dropdown(options=['False', 0.4, 0.6, 0.8], value='False', description='SMOTENC')
one_hot_widget = widgets.Dropdown(options=['label_enc', 'one_hot'], value='label_enc', description='one_hot')
Dropout_widget = widgets.Dropdown(options=[0.00, 0.15, 0.25, 0.4], value=0.00, description='Dropout')
BatchNorm_widget = widgets.Checkbox(value=False, description='BatchNorm')

@interact(
    isol_forest=widgets.Checkbox(value=True, description='isol_forest'),
    iqr=widgets.Checkbox(value=True, description='iqr'),
    box_cox=widgets.Checkbox(value=True, description='box_cox'),
    robust_scaler=widgets.Checkbox(value=True, description='robust_scaler'),
    normalization=widgets.Dropdown(
        options=['False', 'normalization', 'standardization'], value='False', description='normalization'
    ),
    SMOTENC=widgets.Dropdown(
        options=['False', 0.4, 0.6, 0.8], value='False', description='SMOTENC'
    ),
    one_hot=widgets.Dropdown(
        options=['label_enc', 'one_hot'], value='label_enc', description='one_hot'
    ),
    Dropout=widgets.Dropdown(
        options=[0.00, 0.15, 0.25, 0.4], value=0.00, description='Dropout'
    ),
    BatchNorm=widgets.Checkbox(value=True, description='BatchNorm')
)
def update_plot(isol_forest, iqr, box_cox, robust_scaler, normalization, SMOTENC, one_hot, Dropout, BatchNorm):
    filtered_row = data[
        (data['isol_forest'] == isol_forest) &
        (data['iqr'] == iqr) &
        (data['box_cox'] == box_cox) &
        (data['robust_scaler'] == robust_scaler) &
        (data['normalization'] == normalization) &
        (data['SMOTENC'] == SMOTENC) &
        (data['one_hot'] == one_hot) &
        (data['Dropout'] == Dropout) &
        (data['BatchNorm'] == BatchNorm)
    ]
    
    if filtered_row.empty:
        print("No matching data for the selected combination.")
        return
    
    history = filtered_row['history'].iloc[0]
    plot_history(history)
    tsne_encoded = filtered_row['t-SNE'].iloc[0]
    cm_encoded = filtered_row['confusion_matrix'].iloc[0]
    plot_images(tsne_encoded, cm_encoded)

```
